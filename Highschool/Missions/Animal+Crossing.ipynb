```python
# Task 1.1
class AC_Object:
    def __init__(self, name, symbol):
        self.name = name
        self.symbol = symbol

    def get_name(self):
        return self.name
    
    def __str__(self):
        return self.symbol

class Plant(AC_Object):
    def __init__(self, name, type, symbol, days_to_mature):
        super().__init__(name, symbol)
        self.type = type
        self.days_to_mature = days_to_mature
        self.day_timer = 0

    def get_type(self):
        return self.type

    def get_day_timer(self):
        return self.day_timer
    
    def pass_one_day(self):
        self.day_timer += 1

    def check_mature(self):
        return self.day_timer >= self.days_to_mature
    
    def mature(self):
        if not self.check_mature():
            return None
        if self.type == 'flower':
            self.symbol = '¥'
            return None
        if self.type == 'tree':
            return [AC_Object('Apple', '@'), AC_Object('Apple', '@'), AC_Object('Apple', '@')]
        
    def __str__(self):
        return self.symbol
```

```python
# Task 1.2
class Map:
    def __init__(self, max_row, max_col):
        self.max_row = max_row-1
        self.max_col = max_col-1
        self.map = [['' for i in range(self.max_col)] for j in range(self.max_row)]
        self.plants = []

    def add_new_obj(self, obj, row, col):
        if self.map[row][col] != '':
            return print('Unable to add object. The position is already occupied by {}'.format(self.map[row][col]))
        if type(obj) == Plant:
            self.plants.append([obj, row, col])

        for i in self.plants:
            if obj.get_type() == 'tree':
                if (row >= 2 and col >= 2) or (row >= 2 and col <= self.max_col-2) or (row <= self.max_row-2 and col >= 2) or (row <= self.max_row-2 and col <= self.max_col-2):
                    if self.map[row-1][col-1] == '' and self.map[row-1][col] == '' and self.map[row-1][col+1] == '' and self.map[row][col-1] == '' and self.map[row][col+1] == '' and self.map[row+1][col-1] == '' and self.map[row+1][col] == '' and self.map[row+1][col+1] == '':
                        self.map[row][col] = obj
                        return
                    else:
                        return 'Unable to add object. There is another object in the surrounding 3x3'
        self.map[row][col] = obj
    

    def display(self):
        border = "+{}+".format("-" * self.max_col)
        print(border)
        for row in self.map:
            line = "|"
            for obj in row:
                if obj == '':
                    line += " "
                else:
                    line += obj.__str__()
            line += "|"
            print(line)
        print(border)

    def pass_a_day(self):
        for pln in self.plants:
            pln[0].pass_one_day()
            if pln[0].check_mature():
                if pln[0].get_type() == 'tree':
                    child = pln[0].mature()
                    self.map[pln[1]][pln[2]-1] = child[0]
                    self.map[pln[1]][pln[2]+1] = child[1]
                    self.map[pln[1]-1][pln[2]] = child[2]
                    
                else:
                    pln[0].mature()
                    self.map[pln[1]][pln[2]] = pln[0]
                

```
