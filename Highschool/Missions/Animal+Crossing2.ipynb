```python
# Task 1.1
class AC_Object:
    def __init__(self, name, symbol):
        self._name = name
        self._symbol = symbol
    
    def get_name(self):
        return self._name
    
    def __str__(self):
        return self._symbol
    
class Plant(AC_Object):
    def __init__(self, name, type, symbol, days_to_mature):
        super().__init__(name, symbol)
        self._type = type
        self._days_to_mature = days_to_mature-1
        self._day_timer = 0

    def get_type(self):
        return self._type
    
    def get_day_timer(self):
        return self._day_timer
    
    def pass_one_day(self):
        self._day_timer += 1

    def check_mature(self):
        if self._day_timer >= self._days_to_mature:
            return True
        return False
    
    def mature(self):
        if self.check_mature():
            if self._type == "flower":
                self._symbol = "짜"
                return True
            elif self._type == "tree":
                return [AC_Object('Apple', '@'), AC_Object('Apple', '@'), AC_Object('Apple', '@')]
        
        else:
            return None
```

```python
# Task 1.2
class Map:
    def __init__(self, col, row):
        self._col = col-1
        self._row = row-1
        self._map = [[' ' for i in range(self._row)] for i in range(self._col)]
        self._obj_store = []

    def add_new_obj(self, new_obj, col, row):
        if self._map[col][row] == ' ':
            try:
                if new_obj.get_type == 'tree':
                    try:
                        if (self._map[col][row-1] == ' ') and (self._map[col+1][row] == ' ') and (self._map[col+1][row] == ' '):
                            self._map[col][row] = new_obj
                            self._obj_store.append([new_obj, col, row])
                        else:
                            return "Uable to add object, as surrounding tiles is occupied"
                    except:
                        return "Unable to add object, as too close to border"
                else:
                    if self._map[col][row] == ' ':
                        self._map[col][row] = new_obj
                        self._obj_store.append([new_obj, col, row])
                    else:
                        return "Unable to add object, as tile is occupied by {}".format(self._map[row][col])
            except:
                if self._map[col][row] == ' ':
                        self._map[col][row] = new_obj
                else:
                    return "Unable to add object, as tile is occupied by {}".format(self._map[row][col])

    def display(self):
        border = "+{}+".format("-" * self._row)
        print(border)
        for col in self._map:
            line = "|"
            for obj in col:
                if obj == '':
                    line += " "
                else:
                    line += obj.__str__()
            line += "|"
            print(line)
        print(border)
    
    def pass_a_day(self):
        for obj in self._obj_store:
            
            obj[0].pass_one_day()
            if obj[0].check_mature():
                if obj[0].get_type() == 'tree':
                    print(obj[0]._days_to_mature==obj[0]._day_timer)
                    child = obj[0].mature()
                    self._map[obj[1]-1][obj[2]] = child[0]
                    self._map[obj[1]][obj[2]-1] = child[1]
                    self._map[obj[1]+1][obj[2]] = child[2]
                    
                else:
                    obj[0].mature()
                    self._map[obj[1]][obj[2]] = obj[0]
                    
```
